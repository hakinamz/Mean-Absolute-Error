```python
train_path = '../input/food-prices-in-turkey/train.csv'
test_path= '../input/food-prices-in-turkey/test.csv' 
```

```python
df = pd.read_csv(train_path)
```

```python
df
```

```python
df.Place.unique()
```

```python
df.Place = df.Place.replace('National Average',0)
df.Place = df.Place.replace('Ankara',1)
df.Place = df.Place.replace('Istanbul',2)
df.Place = df.Place.replace('Izmir',3)
```

```python
df.Place.unique()
```

```python
df = df.drop(['ProductName', 'UmName','Month'], axis=1)
```

```python
def z_score_normalize(X):
    m = X.shape[0]
    n = 1
    for i in range(n):
        X = (X - X.mean(axis=0))/X.std(axis=0)
    return X
```

```python
Y = df['Price']
X = df.drop(['Price'], axis=1)
```

```python
Y = z_score_normalize(Y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
model= LinearRegression()
```

```python
model.fit(X, Y)
```

```python
test_data = pd.read_csv(test_path)
```

```python
test_data = test_data.drop(['ProductName', 'UmName','Month'], axis=1)
```

```python
test_data.Place = df.Place.replace('National Average',0)
test_data.Place = df.Place.replace('Ankara',1)
test_data.Place = df.Place.replace('Istanbul',2)
test_data.Place = df.Place.replace('Izmir',3)
```

```python
y = test_data['Price']
x = test_data.drop(['Price'],axis = 1)
```

```python
y.describe()
```

```python
y = z_score_normalize(y)
```

```python
x
```

```python
preds = model.predict(x)
```

```python
mean_absolute_error(y, preds)
```
